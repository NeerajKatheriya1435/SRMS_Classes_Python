### BREAK AND CONTINUE STATEMENTS IN PYTHON

```python
# Break Statements

for i in range(1, 10):
    if i == 5:
        break
    print(i)

# Continue Statements

for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

### LIST IN PYTHON

```python
# Homogeneous Data
lst1 = [1,2,2,3,5,4,6]
lst2 = ["Red", "Green", "Blue"]
print(lst1)
print(lst2)

# heterogeneous Data
details = ["Abhijeet", 18, "FYBScIT", 9.8]
print(details) 

# BASICS OF LIST INDEXING

my_list = ['apple', 'banana', 'cherry']

# Accessing Element
print(my_list[0])
print(my_list[-1])

# Update List
# my_list[1] = 'mango'
print(my_list)

# Using Index in Loops
for i in range(len(my_list)):
    print(f"Element at index {i} is {my_list[i]}")

# List Slicing (a related concept)
print(my_list[0:2])   # ['apple', 'mango'] â€” from index 0 to 1
print(my_list[:])     # full list
print(my_list[::2])   # skip every other element

# LIST COMPREHENSION
names = ["Milo", "Sarah", "Bruno", "Anastasia", "Rosa"]
namesWith_O = [item for item in names if "o" in item]
print(namesWith_O)
```

```python
### LIST METHODS

colors = ["voilet", "indigo", "blue", "green"]
# To sort list
colors.sort()
print(colors)
num = [4,2,5,3,6,1,2,1,2,8,9,7]
num.sort()
print(num)

num = [4,2,5,3,6,1,2]
# To reverse list
num.reverse()
print(num)

# Access element at specific index
print(num.index(3))

# Count
colors = ["voilet", "green", "indigo", "blue", "green"]
print(colors.count("green"))

# Copy
colors = ["voilet", "green", "indigo", "blue"]
newlist = colors.copy()
print(colors)
print(newlist)

# Append
colors = ["voilet", "indigo", "blue"]
colors.append("green")
print(colors)
```

### TUPLES IN PYTHON

```python
tuple1 = (1,2,2,3,5,4,6)
tuple2 = ("Red", "Green", "Blue")

# Accessing tuple items:
print(tuple1)
print(tuple2[1])

# Check for item:
country = ("Spain", "Italy", "India", "England", "Germany")
if "Germany" in country:
    print("Germany is present.")
else:
    print("Germany is absent.")

# Range of Index:
animals = ("cat", "dog", "bat", "mouse", "pig", "horse", "donkey", "goat", "cow")
print(animals[3:7])     #using positive indexes
print(animals[-7:-2])   #using negative indexes

# Manipulating Tuples
countries = ("Spain", "Italy", "India", "England", "Germany")
temp = list(countries)
temp.append("Russia")       #add item 
temp.pop(3)                 #remove item
temp[2] = "Finland"         #change item
countries = tuple(temp)
print(countries)
```

### STRING FORMATTING IN PYTHON

```python
# 1. Using format() method (Python 3 and above)
name = "Alice"
age = 25
print("My name is {} and I am {} years old.".format(name, age))

# With index numbers:
print("My name is {0} and I am {1} years old. {0} is learning Python.".format(name, age))

# With named placeholders:
print("My name is {n} and I am {a} years old.".format(n="Alice", a=25))

# 2. Using f-strings (Python 3.6+)
name = "Bob"
age = 30
print(f"My name is {name} and I am {age} years old.")
```

### WHAT IS A SET IN PYTHON?

```python
my_set = {1, 2, 3, 4}
another_set = set([3, 4, 5])

# Opeartion in set

a = {1, 2, 3}
b = {2, 3, 4}
print(a | b) # Union: {1, 2, 3, 4}
print(a & b) # Intersection: {2, 3}
print(a - b) # Difference: {1}
print(a ^ b) # Symmetric Difference: {1, 4}

# SET MEMBERSHIP

if 2 in my_set:
    print("2 is present")

if 9 not in my_set:
    print("9 is not present")
```

### WHAT IS A DICTIONARY IN PYTHON?

```python
# CREATING A DICTIONARY:

# Basic dictionary
person = {'name': 'Alice', 'age': 25}

# Empty dictionary
empty_dict = {}

# ACCESSING AND MODIFYING ITEMS:

# Access value by key
print(person['name'])         # Output: Alice

# Add or update a value
person['age'] = 26            # Updates age
person['city'] = 'London'     # Adds new key-value pair

# Using get() (returns None if key doesn't exist)
print(person.get('gender'))   # Output: None

# LOOPING THROUGH A DICTIONARY:

# Loop through keys
for key in person:
    print(key)

# Loop through values
for value in person.values():
    print(value)

# Loop through key-value pairs
for key, value in person.items():
    print(key, value)
```

### HOW TO RAISE A CUSTOM ERROR

```python
age = 12
if age < 0:
    raise ValueError("Age cannot be negative")

# Example
try:
    num = int(input("Enter a positive number: "))
    if num <= 0:
        raise ValueError("Number must be greater than zero.")
    print("You entered:", num)
except ValueError as e:
    print("Error:", e)
```

### SHORT-HAND IF-ELSE IN PYTHON

```python
age = 20
status = "Adult" if age >= 18 else "Minor" 
print(status)

# Example 2
marks = 85

grade = (
    "A+" if marks >= 90 else
    "A" if marks >= 80 else
    "B" if marks >= 70 else
    "C" if marks >= 60 else
    "Fail"
)

print("Grade:", grade)
```
