### IF-ELSE STATEMENTS

```python

# if-else statements

age = 21
age_normal = 18
if (age> age_normal):
    print("You can drive car.")
else:
    print("You can not drive car.")

# if-else with elif
marks = 75
if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
else:
    print("Fail")
```

### MATCH CASE STATEMENT

```python
# Exmaple 1
value = 2
match value:
    case 1:
        print("Between 1 and 3")
    case 4:
        print("It's four")
    case _:
        print("Something else")

# Exmaple 2
value = 2
match value:
    case 1 | 2 | 3:
        print("Between 1 and 3")
    case 4:
        print("It's four")
    case _:
        print("Something else")
    
# Example 3
number = 10
match number:
    case 1 | 2 | 3:
        print("Between 1 and 3")
    case 4:
        print("It's four")
    case _ if number % 2 == 0:
        print("Even number")
    case _:
        print("Odd number")
```

### Strings in python

```python
name = "Ayush"
print("Hello, " + name)

# Multiline Strings:
a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a)

print("The character at index 0:",name[0])
print("The character at index 1:",name[1])

# Iterate whole string using for loop
for character in name:
    print(character)

# STRING SLICING & OPERATIONS ON STRING
fruit = "Shivam"
len1 = len(fruit)
print("Shivam is a", len1, "letter word.")

pie = "ApplePie"
print(pie[:5])      #Slicing from Start
print(pie[5:])      #Slicing till End
print(pie[2:6])     #Slicing in between
print(pie[-8:])     #Slicing using negative index
```

### BREAK AND CONTINUE STATEMENTS IN PYTHON

```python
# Break Statements

for i in range(1, 10):
    if i == 5:
        break
    print(i)

# Continue Statements

for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

### LIST IN PYTHON

```python
# Homogeneous Data
lst1 = [1,2,2,3,5,4,6]
lst2 = ["Red", "Green", "Blue"]
print(lst1)
print(lst2)

# heterogeneous Data
details = ["Abhijeet", 18, "FYBScIT", 9.8]
print(details) 

# BASICS OF LIST INDEXING

my_list = ['apple', 'banana', 'cherry']

# Accessing Element
print(my_list[0])
print(my_list[-1])

# Update List
# my_list[1] = 'mango'
print(my_list)

# Using Index in Loops
for i in range(len(my_list)):
    print(f"Element at index {i} is {my_list[i]}")

# List Slicing (a related concept)
print(my_list[0:2])   # ['apple', 'mango'] â€” from index 0 to 1
print(my_list[:])     # full list
print(my_list[::2])   # skip every other element

# LIST COMPREHENSION
names = ["Milo", "Sarah", "Bruno", "Anastasia", "Rosa"]
namesWith_O = [item for item in names if "o" in item]
print(namesWith_O)
```

```python
### LIST METHODS

colors = ["voilet", "indigo", "blue", "green"]
# To sort list
colors.sort()
print(colors)
num = [4,2,5,3,6,1,2,1,2,8,9,7]
num.sort()
print(num)

num = [4,2,5,3,6,1,2]
# To reverse list
num.reverse()
print(num)

# Access element at specific index
print(num.index(3))

# Count
colors = ["voilet", "green", "indigo", "blue", "green"]
print(colors.count("green"))

# Copy
colors = ["voilet", "green", "indigo", "blue"]
newlist = colors.copy()
print(colors)
print(newlist)

# Append
colors = ["voilet", "indigo", "blue"]
colors.append("green")
print(colors)
```

### STRING FORMATTING IN PYTHON

```python
# 1. Using format() method (Python 3 and above)
name = "Alice"
age = 25
print("My name is {} and I am {} years old.".format(name, age))

# With index numbers:
print("My name is {0} and I am {1} years old. {0} is learning Python.".format(name, age))

# With named placeholders:
print("My name is {n} and I am {a} years old.".format(n="Alice", a=25))

# 2. Using f-strings (Python 3.6+)
name = "Bob"
age = 30
print(f"My name is {name} and I am {age} years old.")
```

### SHORT-HAND IF-ELSE IN PYTHON

```python
age = 20
status = "Adult" if age >= 18 else "Minor" 
print(status)

# Example 2
marks = 85

grade = (
    "A+" if marks >= 90 else
    "A" if marks >= 80 else
    "B" if marks >= 70 else
    "C" if marks >= 60 else
    "Fail"
)

print("Grade:", grade)
```
