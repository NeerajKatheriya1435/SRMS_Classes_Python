```python
import numpy as np
arr = np.array([1,2,3,4,5,6,7])
print(arr[1:3])
```

```python
print(arr[:4])
```

```python
print(arr[1:])
```

```python
print(arr[1:6:2])
```

```python
print(arr[-3:-1])
```

```python
print(arr[::2])
```

```python
arr2 = np.array([[1,2,3,4],[5,6,7,8]])
print(arr2[1, 1:3])
```

```python
print(arr2[0:2, 2])
```

```python
print(arr2[0:2,1:3])
```
