### Data Analysis Using Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structures : [Restaurent_id, year 2021, year 2022, year 2023, year 2024]
salesData=np.array([
    [101,2000,4000,3000,8000,5000],
    [102,4000,7000,3000,8000,3000],
    [103,6000,4000,3000,3000,5000],
    [104,2000,3000,6000,8000,7000],
    [105,3000,7000,3000,8000,6000]
    ])
```

```python
# Print first three rows
print("First Three Rows: ",salesData[0:3])

#all the sales in 2021 of every item
print("First Column: ",salesData[:,1])

#all the sales in all year if every item except First Column
print("Except First Column: ",salesData[:,1:])
```

```python
print("Total Sales Per Year: ",np.sum(salesData[:,1:],axis=0))
print("Mininum Sales Of Restaurant (Row): ",np.min(salesData[:,1:],axis=1))
print("Maximum Sales Per Year (Column): ",np.max(salesData[:,1:],axis=0))
print("Average Sales Per Restaurant (Row): ",np.mean(salesData[:,1:],axis=0))
print("Cumlative Sales Per Restaurant (Row): ",np.cumsum(salesData[:,1:],axis=1))
```

### Chart Representation using matlplotlib

```python
plt.figure(figsize=(8,6))
plt.plot(np.mean(np.cumsum(salesData[:,1:],axis=1),axis=0))
plt.title("Average Sales Per Year")
plt.xlabel("Years")
plt.ylabel("Sales Data")
plt.grid(True)
plt.show()
```

### Vectors Operations

```python
myArr1=np.array([1,5,9,3,6])
myArr2=np.array([6,8,9,2,5])

print("Addition of Vector: ",(myArr1+myArr2))
print("Multiplication of Vector: ",(myArr1*myArr2))
print("Dot of Vector: ",np.dot(myArr1,myArr2))

myAngle=np.arccos(np.dot(myArr1,myArr2)/((np.linalg.norm(myArr1)*(np.linalg.norm(myArr2)))))
print("Angle: ",myAngle)
```

```python
myStd=np.array(["Shubham","Shiva","Vandana","Sahu","Khushi"])
toUpperCase=np.vectorize(str.upper)
print("Upper Case: ",toUpperCase(myStd))
```
