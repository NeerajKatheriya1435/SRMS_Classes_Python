### Random Array Creation

```python
var1=np.random.rand(5)
var2=np.random.rand(3,3)
var3=np.random.randn(5)
var4=np.random.randn(3,4)
var5=np.random.ranf(5)
var6=np.random.ranf((4,5))
var7=np.random.randint(4,10,6)
print("Random 1d Array Between 0 to 1:",var1)
print("Random 2d Array Between 0 to 1:",var2)
print("Random 1d Array Between 0 to 1 Negative Positive:",var3)
print("Random 2d Array Between 0 to 1 Negative Positive:",var4)
print("Random 1d Array Between 0 to 1:",var5)
print("Random 2d Array Between 0 to 1:",var6)
print("Random 1d Array Between 0 to 1:",var7)
```
