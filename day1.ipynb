### TYPECASTING IN PYTHON

```python
print("6"+"9")
print(6+9.7)

str1=int("6")
str2=int("9")
print(str1+str2)
```

```python
# 1. IMPLICIT TYPECASTING

a = 5       # int
b = 2.0     # float
result=a+b
print(type(result))  # Output: <class 'float'>

# 2. EXPLICIT TYPECASTING

# String to int
num_str = "10" #print(“Hello” 6,9,0,sep=”$”,end=”JJA”)
num_int = int(num_str)
print(num_int)
print(type(num_int))         # Output: 10

# Float to int
f = 7.9
print(int(f))          # Output: 7 (decimal part is removed)
print(type(int(f))) 
```

### Operators in Python

```python
# 1. Arithmetic Operators:

a=14
b=3
print("Addition:", a + b)        # 13
print("Subtraction:", a - b)     # 7
print("Multiplication:", a * b)  # 30
print("Division:", a / b)        # 3.333...
print("Floor Division:", a // b) # 3
print("Modulus:", a % b)         # 1
print("Exponent:", a ** b)       # 1000
print("-----------------------------------")

# 2. Assignment Operators
# 6==6
# a=6

x=8
# x=x+3
x += 3   # x = x + 3 (11)
print("x += 3 →", x)  # 8
x -= 2   # x = x - 2  (9)
print("x -= 2 →", x)  # 6
x *= 4   # x = x * 4
print("x *= 4 →", x)  # 24
x /= 6   # x = x / 6
print("x /= 6 →", x)  # 4.0
x //= 2  # x = x // 2
print("x //= 2 →", x) # 2.0
x **= 3  # x = x ** 3
print("x **= 3 →", x) # 8.0
print("-----------------------------------")
```

```python
# 3.Comparison Operators

p = 8
q = 5
print("p == q:", p == q)   # False
print("p != q:", p != q)   # True
print("p > q:", p > q)     # True
print("p < q:", p < q)     # False
print("p >= q:", p >= q)   # True
print("p <= q:", p <= q)   # False
print("-----------------------------------")

# 4.Bitwise Operators

a = 5   # 0101 in binary  010  >01010
b = 3   # 0011 in binary
print("a & b:", a & b)     # 1 (0001)
print("a | b:", a | b)     # 7 (0111)
print("a ^ b:", a ^ b)     # 6 (0110)
print("a << 1:", a << 1)   # 10 (1010)
print("a >> 1:", a >> 1)   # 2 (0010)
print("-----------------------------------")

# 0101
# ----
# 01010
# -----
#   0101
# ----
#    010
```

```python
# 6.Membership Operators

# l1=[6,"aman","suman","abhay"]

list1 = [1, 2, 3, 4, 5]
print("Membership Operators:")
print("3 in list1:", 38 in list1)       # True
print("7 not in list1:", 778 not in list1) # True
print("-----------------------------------")

# 7.Identity Operators

a = 10
b = 10
c = [1, 2]
d = [1, 2]
print("Identity Operators:")
print("a is b:", a is b)     # True (same memory for small ints (-5 to 255))
print("c is d:", c is d)     # False (different memory)
print("c == d:", c == d)     # True (values are same)
print("a is not b:", a is not b)  # False
print("-----------------------------------")
```
